```python
#This script will perform statistical functions and the Kolmogorov-Smirnov test to...
#determine if certain variables such a muscovite, biotite and porosity for all outcrops come...
#from the same distribution.
#import pandas in order to perform statistical functions
import pandas as pd
#import statistics for kurtosis and skewness functions
from scipy.stats import kurtosis, skew
#import matplotlib to plot histograms
import matplotlib.pyplot as plt
#import statsmodel for Kolmogorov-Smirnov test to determine if the samples came from the same distribution
from scipy.stats import ks_2samp

#import thinsection data from excel 
thinSectionData = pd.read_excel('/Users/sarahcronin/Desktop/Testdata.xlsx', sheet_name = 'Sheet2');
#set indexing values
thinSectionData.set_index('Samples', inplace=True);
print(thinSectionData)
#calculate and print the mean
dMean = thinSectionData.loc["Muscovite", :].mean();
print(dMean);
#calculate and print the meadian
dMedian = thinSectionData.loc["Muscovite", :].median();
print(dMedian)
#calculate and print the mode
dMode = thinSectionData.loc["Muscovite", :].mode();
#dMode = float(dMode);
print(dMode)
#calculate and print the standard deviation
dStdDev = thinSectionData.loc["Muscovite", :].std();
print(dStdDev)
#calculate and print the variance
dVar = thinSectionData.loc["Muscovite", :].var();
print(dVar)
#calculate and print the skewness
dSkew = skew(thinSectionData.loc["Muscovite", :]);
print(dSkew)
#calculate and print the kurtosis
dKurt= kurtosis(thinSectionData.loc["Muscovite", :]);
print(dKurt)
#perform a Kolmogorov-Smirnov test
KsTestResult = ks_2samp(thinSectionData.loc["Muscovite", 'N-03':'N-04'], thinSectionData.loc["Muscovite", 'N-07':'N-09'])
#create a histogram for muscovite percentage
fig, axs = plt.subplots(1, 1, figsize =(10,7), tight_layout = True)
axs.hist(thinSectionData.loc["Muscovite", :])
plt.xlabel("Percentage")
plt.ylabel("Number")
plt.title("Muscovite Percentage")
#show plot
plt.show()

#calculate and print the mean
dMean1 = thinSectionData.loc["Biotite", :].mean();
print(dMean1);
#calculate and print the meadian
dMedian1 = thinSectionData.loc["Biotite", :].median();
print(dMedian1)
#calculate and print the mode
dMode1 = thinSectionData.loc["Biotite", :].mode();
#dMode1 = float(dMode1);
print(dMode1)
#calculate and print the standard deviation
dStdDev1 = thinSectionData.loc["Biotite", :].std();
print(dStdDev1)
#calculate and print the variance
dVar1 = thinSectionData.loc["Biotite", :].var();
print(dVar1)
#calculate and print the skewness
dSkew1 = skew(thinSectionData.loc["Biotite", :]);
print(dSkew1)
#calculate and print the kurtosis
dKurt1 = kurtosis(thinSectionData.loc["Biotite", :]);
print(dKurt1)
#perform a Kolmogorov-Smirnov test
KsTestResult1 = ks_2samp(thinSectionData.loc["Biotite", 'N-03':'N-04'], thinSectionData.loc["Biotite", 'N-07':'N-09'])
#create a histogram for biotite percentage
fig, axs = plt.subplots(1, 1, figsize =(10,7), tight_layout = True)
axs.hist(thinSectionData.loc["Biotite", :])
plt.xlabel("Percentage")
plt.ylabel("Number")
plt.title("Biotite Percentage")
#show plot
plt.show()

#calculate and print the mean
dMean2 = thinSectionData.loc["Holes", :].mean();
print(dMean2);
#calculate and print the meadian
dMedian2 = thinSectionData.loc["Holes", :].median();
print(dMedian2)
#calculate and print the mode
dMode2 = thinSectionData.loc["Holes", :].mode();
#dMode2 = float(dMode2);
print(dMode2)
#calculate and print the standard deviation
dStdDev2 = thinSectionData.loc["Holes", :].std();
print(dStdDev2)
#calculate and print the variance
dVar2 = thinSectionData.loc["Holes", :].var();
print(dVar2)
#calculate and print the skewness
dSkew2 = skew(thinSectionData.loc["Holes", :]);
print(dSkew2)
#calculate and print the kurtosis
dKurt2 = kurtosis(thinSectionData.loc["Holes", :]);
print(dKurt2)
#perform a Kolmogorov-Smirnov test
KsTestResult2 = ks_2samp(thinSectionData.loc["Holes", 'N-03':'N-04'], thinSectionData.loc["Holes", 'N-07':'N-09'])
#create a histogram for porosity(holes) percentage
fig, axs = plt.subplots(1, 1, figsize =(10,7), tight_layout = True)
axs.hist(thinSectionData.loc["Holes", :])
plt.xlabel("Percentage")
plt.ylabel("Number")
plt.title("Porosity")
#show plot
plt.show()
```
